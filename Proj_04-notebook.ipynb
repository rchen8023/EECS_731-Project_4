```python
import pandas as pd

data = pd.read_csv('Data/mlb_elo.csv')
data.shape
```

There are 220946 rows in the data, I first tired to drop all Nan values, however, data reduced to 1580 rows after drop all Nan value, it is too small to prediction. And I noticed that most Nan value are from feature 'playoff', so I removed column 'playoff', then drop all other Nan values. 

```python
data = data.drop('playoff', axis=1)
data = data.dropna()
data
```

In this datasets, there are 4 columns in strings, 'team1', 'team2', 'pitcher1', 'pitcher2'. I then encoded this 4 columns by using LabelEncoder from sklearn. And delete a column 'date' which is not used.

```python
from sklearn import preprocessing
team_encodingColumns = data['team1'].append(data['team2'])
pitcher_encodingColumns = data['pitcher1'].append(data['pitcher2'])

team_encoding = preprocessing.LabelEncoder()
pitcher_encoding = preprocessing.LabelEncoder()
team_encoding.fit(team_encodingColumns)
pitcher_encoding.fit(pitcher_encodingColumns)

data['team1'] = team_encoding.transform(data['team1'])
data['team2'] = team_encoding.transform(data['team2'])
data['pitcher1'] = pitcher_encoding.transform(data['pitcher1'])
data['pitcher2'] = pitcher_encoding.transform(data['pitcher2'])

data = data.drop('date',axis=1)
data.head()
```

First split the data to training data and testing data with seperate sample and labels. Labels are two scores. 

```python
from sklearn.model_selection import train_test_split
label = data[['score1','score2']]
sample = data.drop(['score1','score2'],axis=1)
sample_train, sample_test, label_train, label_test = train_test_split(sample, label, test_size=0.1)
```

I first tried the Linear Regression model to make the prediction.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

LR = LinearRegression()
LR.fit(sample_train,label_train)
LR_predict = LR.predict(sample_test)
accuracy_LR = r2_score(label_test,LR_predict)
accuracy_LR
```

The accuracy of this model is very low, since there is no parameters we can change to improve the accuracy, so then I tried the Random Forest Regression model.

```python
from sklearn.ensemble import RandomForestRegressor

RF = RandomForestRegressor(n_estimators=10)
RF.fit(sample_train,label_train)
RF_predict = RF.predict(sample_test)
accuracy_RF = r2_score(label_test,RF_predict)
accuracy_RF
```

The accuracy for this model is also very low, then I tried to change some of the parameters to improve the accuracy. 

At first, I planed to try to change the maximun depth of random tree. Since when the tree keep expand untill all leaves very pure, then it may have a case that have some leaves with only few samples, and that may cause overfitting. However, I still cannot find a way to figure out what is the final depth when the leaves are all pure, so I did not know which range of depth I can choose. In another way, when the depth of the tree is decreased, even overfitting does not happen, the tree may not pure enough to make better prediction accuracy. 

Then I tried to add more random trees to make the model more complex.

```python
RF_1 = RandomForestRegressor(n_estimators=20)
RF_1.fit(sample_train,label_train)
RF_predict_1 = RF_1.predict(sample_test)
accuracy_RF1 = r2_score(label_test,RF_predict_1)
accuracy_RF1
```

```python
RF_2 = RandomForestRegressor(n_estimators=50)
RF_2.fit(sample_train,label_train)
RF_predict_2 = RF_2.predict(sample_test)
accuracy_RF2 = r2_score(label_test,RF_predict_2)
accuracy_RF2
```

```python
RF_3 = RandomForestRegressor(n_estimators=100)
RF_3.fit(sample_train,label_train)
RF_predict_3 = RF_3.predict(sample_test)
accuracy_RF3 = r2_score(label_test,RF_predict_3)
accuracy_RF3
```

```python
RF_4 = RandomForestRegressor(n_estimators=200)
RF_4.fit(sample_train,label_train)
RF_predict_4 = RF_4.predict(sample_test)
accuracy_RF4 = r2_score(label_test,RF_predict_4)
accuracy_RF4
```

As the result, the accuracy will increasing while more random tree is builded, but the implementation would be very slow and the accuracy only have few improve. 

Then I also tried with Gradient Boost and Neural Network meodel

The Gradient Boost only take one column of label value, so I make two model for each scores. 

```python
from sklearn.ensemble import GradientBoostingRegressor

GB1 = GradientBoostingRegressor()
GB1.fit(sample_train,label_train['score1'])
GB_predict1 = GB1.predict(sample_test)
accuracy_GB_1 = r2_score(label_test['score1'],GB_predict1)
print(accuracy_GB_1)

GB2 = GradientBoostingRegressor()
GB2.fit(sample_train,label_train['score2'])
GB_predict2 = GB2.predict(sample_test)
accuracy_GB_2 = r2_score(label_test['score2'],GB_predict2)
print(accuracy_GB_2)
```

```python
from sklearn.neural_network import MLPRegressor

MLP = MLPRegressor(max_iter=500)
MLP.fit(sample_train,label_train)
MLP_predict = MLP.predict(sample_test)
accuracy_MLP = r2_score(label_test,MLP_predict)
accuracy_MLP
```

The result of both Gradient Boost and Neural Network are not have a better performance than the Random Forest, but when Random Forest has better accuracy, it implementing very slow.
